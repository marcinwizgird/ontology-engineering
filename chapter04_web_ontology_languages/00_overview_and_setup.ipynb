## Tooling
The book uses **Protégé + HermiT/Pellet** (Java). To keep everything runnable in Python with no Java, we use:

- **owlready2** — Pythonic OWL (classes, properties, restrictions, serialisation).
- **rdflib** — RDF graphs, serialisation, SPARQL.
- **owlrl** — pure-Python **OWL 2 RL / RDFS** reasoner for the inference demos.

> DL classification (e.g. HermiT) needs Java; OWL 2 RL forward-chaining does not, so 'run the reasoner' works anywhere. Where full DL reasoning matters, we say so.
