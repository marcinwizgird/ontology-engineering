## Exercise 4.2 — injury to a part is an injury to the whole
Model: an injury to a bone in your hand is an injury to your hand, generalised to any anatomical part→whole. **Which OWL 2 DL feature?** A **property chain**: `injuryOf ∘ partOf ⊑ injuryOf`. Formalise and let the reasoner propagate it.

```python
# SOLUTION KEY — property chain + OWL 2 RL inference
from owlready2 import Thing, ObjectProperty, PropertyChain
w = ch4.new_world(); onto = w.get_ontology('http://example.org/anat#')
with onto:
    class BodyPart(Thing): pass
    class Injury(Thing): pass
    class partOf(ObjectProperty): pass
    class injuryOf(ObjectProperty): pass  # Injury -> BodyPart
    injuryOf.property_chain = [PropertyChain([injuryOf, partOf])]
    hand = BodyPart('hand'); bone = BodyPart('handBone'); cut = Injury('cut1')
    bone.partOf = [hand]; cut.injuryOf = [bone]
g, b, a, new = ch4.reason_owlrl(onto)
from rdflib import Namespace
AN = Namespace('http://example.org/anat#')
print('cut1 injuryOf hand inferred:', (AN.cut1, AN.injuryOf, AN.hand) in g)
```

## Exercise 4.5 — unqualified vs qualified cardinality changes the species
(a) `Bicycle ⊑ ≥2 hasComponent.⊤`  →  (b) `Bicycle ⊑ ≥2 hasComponent.Wheel`. Inspect how the feature profile changes.

```python
# SOLUTION KEY
from owlready2 import Thing, ObjectProperty
def bike(qualified):
    w = ch4.new_world(); o = w.get_ontology('http://ex/bike#')
    with o:
        class Wheel(Thing): pass
        class hasComponent(ObjectProperty): pass
        class Bicycle(Thing): pass
        Bicycle.is_a.append(hasComponent.min(2, Wheel) if qualified else hasComponent.min(2, Thing))
    return o
print('UNqualified (≥2 hasComponent.⊤):', ch4.classify_profile(bike(False))['feature_counts'])
print('Qualified  (≥2 hasComponent.Wheel):', ch4.classify_profile(bike(True))['feature_counts'])
print('\nMain difference: the qualified restriction names a filler class (Wheel) — '
      'the Q in SROIQ — which OWL 2 DL allows but OWL 1 DL did not.')
```

## Exercise 4.6 — vegan ⊑ vegetarian?
Build vegan & vegetarian (from Ex. 3.2) and check `O ⊢ Vegan ⊑ Vegetarian` and `O ⊢ Vegetarian ⊑ Vegan`.

```python
# SOLUTION KEY
from owlready2 import Thing, ObjectProperty, Not
w = ch4.new_world(); o = w.get_ontology('http://ex/food#')
with o:
    class Food(Thing): pass
    class AnimalProduct(Food): pass
    class Meat(AnimalProduct): pass   # Meat ⊑ AnimalProduct ⊑ Food
    class eats(ObjectProperty): pass
    class Vegetarian(Thing):
        equivalent_to = [eats.some(Food) & eats.only(Not(Meat))]
    class Vegan(Thing):
        equivalent_to = [eats.some(Food) & eats.only(Not(AnimalProduct))]
# Vegan forbids ALL animal products (incl. meat); Vegetarian only forbids meat.
# Since Meat ⊑ AnimalProduct, ¬AnimalProduct ⊑ ¬Meat, so every Vegan is a
# Vegetarian, but not vice versa.
print('Expected (by definition): Vegan ⊑ Vegetarian = True; Vegetarian ⊑ Vegan = False')
print('Vegan DL     :', ch4.dl_render(o.Vegan))
print('Vegetarian DL:', ch4.dl_render(o.Vegetarian))
print('NOTE: a full proof of the subsumption needs a DL reasoner (HermiT/Java); '
      'the definitions above entail Vegan ⊑ Vegetarian.')
```

## Exercise 4.8 — university ontology: Joint vs Single Honours
(a) `JointHonoursMathsCS` takes **both** CS and Maths modules; (b) `SingleHonoursMaths` takes **only** Maths modules. Possible? Represent it.

```python
# SOLUTION KEY
from owlready2 import Thing, ObjectProperty
w = ch4.new_world(); o = w.get_ontology('http://ex/uni#')
with o:
    class Module(Thing): pass
    class MathsModule(Module): pass
    class CSModule(Module): pass
    class Student(Thing): pass
    class takes(ObjectProperty): pass
    class JointHonoursMathsCS(Student):
        equivalent_to = [Student & takes.some(MathsModule) & takes.some(CSModule)]
    class SingleHonoursMaths(Student):
        equivalent_to = [Student & takes.only(MathsModule) & takes.some(MathsModule)]
print('Joint  :', ch4.dl_render(o.JointHonoursMathsCS))
print('Single :', ch4.dl_render(o.SingleHonoursMaths))
print('\nYes — possible. Joint uses ∃ (someValuesFrom) on both module types; '
      'Single uses ∀ (allValuesFrom) Maths + ∃ Maths so a student takes ONLY Maths.')
```

## Exercise 4.10 — 'exactly 2 modules' and consistency
Add `Student ⊑ =2 takes.Module`. Student 9 takes MT101, CS101, CS102 (3 modules); Student 10 takes MT101, CS101, EC101 (3). Are they consistent?

```python
# SOLUTION KEY (reasoned in Python)
from owlready2 import Thing, ObjectProperty
w = ch4.new_world(); o = w.get_ontology('http://ex/uni2#')
with o:
    class Module(Thing): pass
    class Student(Thing): pass
    class takes(ObjectProperty): pass
    Student.is_a.append(takes.exactly(2, Module))
    s9 = Student('Student9')
    m = [Module('MT101'), Module('CS101'), Module('CS102')]
    s9.takes = m
from owlready2 import AllDifferent
AllDifferent(m)  # WITHOUT this, the reasoner could merge modules to satisfy =2
print('Key insight (no UNA in OWL): with =2 takes.Module and 3 *distinct* modules, '
      'the ontology is INCONSISTENT — but only if the 3 modules are asserted '
      'pairwise different (AllDifferent). Without that, a DL reasoner can infer two '
      'of them are the same individual to keep it consistent.')
print('=> Student 9 (3 distinct modules) violates =2  -> inconsistent.')
print('=> Student 10 likewise (MT101, CS101, EC101 distinct) -> inconsistent.')
```

## Exercise 4.11 — find principal vs knock-on errors
A deliberately broken ontology: distinguish the **principal** unsatisfiability from the **knock-on** ones.

```python
# SOLUTION KEY
from owlready2 import Thing
w = ch4.new_world(); o = w.get_ontology('http://ex/cs#')
with o:
    class Person(Thing): pass
    class Course(Thing): pass
    # PRINCIPAL error: Person and Course disjoint, but AI is declared as both
    from owlready2 import AllDisjoint
    AllDisjoint([Person, Course])
    class AICourse(Person, Course): pass        # principal: unsatisfiable
    class AdvancedAICourse(AICourse): pass      # knock-on: unsat because parent is
print('Principal error : AICourse ⊑ Person ⊓ Course while Person ⊓ Course = ⊥ '
      '(disjoint) -> AICourse is unsatisfiable.')
print('Knock-on error  : AdvancedAICourse is unsatisfiable ONLY because it is a '
      'subclass of AICourse.')
print('Fix: remove one of the disjoint parents from AICourse (keep the knowledge: '
      'a course is not a person); do NOT just delete the unsat classes.')
```

## Exercise 4.12 — AWO: penguins (non-flying birds) & Lepidoptera life stages


```python
# SOLUTION KEY (a) birds & penguins — the classic default/exception modelling issue
from owlready2 import Thing, ObjectProperty
w = ch4.new_world(); o = w.get_ontology('http://ex/awo_birds#')
with o:
    class Animal(Thing): pass
    class Bird(Animal): pass
    class Wing(Thing): pass
    class hasPart(ObjectProperty): pass
    class canFly(ObjectProperty): pass
    # 'birds have wings whose function is to fly' — structural, always true
    Bird.is_a.append(hasPart.some(Wing))
    class Penguin(Bird): pass   # a bird, but does NOT fly
print('Modelling note: OWL has NO non-monotonic defaults. You cannot say '
      '"birds fly" as a default and then make penguins an exception. '
      'Represent the always-true structural fact (Bird ⊑ ∃hasPart.Wing) and model '
      'flying as a separate, non-universal property — or use the function of wings '
      '(Chapter 9/agentive roles), not a blanket "all birds fly".')
print('Penguin DL:', ch4.dl_render(o.Penguin))
```

```python
# SOLUTION KEY (b) Lepidoptera life stages: egg -> larva -> pupa -> adult
from owlready2 import Thing, ObjectProperty
w = ch4.new_world(); o = w.get_ontology('http://ex/lepidoptera#')
with o:
    class Insect(Thing): pass
    class Lepidoptera(Insect): pass     # 'scaled wings' (butterflies & moths)
    class LifeStage(Thing): pass
    class Egg(LifeStage): pass
    class Larva(LifeStage): pass        # caterpillar
    class Pupa(LifeStage): pass         # chrysalis
    class Adult(LifeStage): pass        # butterfly
    class hasStage(ObjectProperty): pass
    Lepidoptera.is_a.append(hasStage.some(Egg))
    Lepidoptera.is_a.append(hasStage.some(Larva))
    Lepidoptera.is_a.append(hasStage.some(Pupa))
    Lepidoptera.is_a.append(hasStage.some(Adult))
print(ch4.dl_render(o.Lepidoptera))
print('Modelling note: the SAME individual is morphologically distinct over time — '
      'a temporal/identity challenge (returns in Chapter 5 & Chapter 10).')
```
